```python
rabbit_years = [
  ["1951-02-06", "1952-01-27"],
  ["1963-01-25", "1964-02-13"],
  ["1975-02-11", "1976-01-31"],
  ["1987-01-29", "1988-02-17"],
  ["1999-02-16", "2000-02-05"]
]
def is_rabbit_year(s: str):
    return any(start <= s < end for start, end in rabbit_years)

def is_cancer_sign(s: str):
    m_d = s[-5:]  # Extract month and day (e.g., "02-06")
    return "06-21" <= m_d <= "07-22"
```
